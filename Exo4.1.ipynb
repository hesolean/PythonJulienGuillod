```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,1,50)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0,2*np.pi,100)

# définir le graphique
plt.figure(figsize=(8,5))
plt.title(r'Fonctions trigonométriques')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$0$', r'$\frac{\pi}{2}$',r'$\pi$',r'$\frac{3\pi}{2}$',
            r'$2\pi$'])

# tracer les courbes
for k in range(1,4):
    plt.plot(x, np.sin(k * x), marker='o', label=r"$\sin({}x)$".format(k))
    plt.plot(x, np.cos(k * x), marker='o', label=r"$\cos({}x)$".format(k))

# Créer la légende en utilisant les lignes stockées dans la liste
plt.legend()



#affichage du graphique
plt.show()
```

```python
# Création d'une matrice bidimensionnelle pour l'exemple
data = np.random.random((10, 10))

# Utilisation de la fonction imshow pour afficher la matrice
plt.imshow(data, cmap='hot', interpolation='bicubic', origin='upper')

# Ajout d'une barre de couleur pour indiquer les valeurs
plt.colorbar()

# Ajout de titres et d'étiquettes d'axe
plt.title('Graphique imshow')
plt.xlabel('Axe X')
plt.ylabel('Axe Y')

# Afficher le graphique
plt.show()
```
